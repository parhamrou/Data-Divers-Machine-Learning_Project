```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
all_data = pd.read_excel('WPP2022.xlsx', skiprows=16, index_col=0)
```

```python
all_data.head(20)
```

```python
all_data.columns
```

```python
countries = all_data[all_data['Type'] == 'Country/Area']
countries
```

```python
country_continent = pd.read_csv('https://gist.githubusercontent.com/stevewithington/20a69c0b6d2ff846ea5d35e5fc47f26c/raw/13716ceb2f22b5643ce5e7039643c86a0e0c6da6/country-and-continent-codes-list-csv.csv')
```

```python
country_continent.to_csv('country_continent.csv')
```

```python
countries = countries.merge(country_continent[['Country_Number', 'Continent_Name', 'Three_Letter_Country_Code']], how='left', left_on='Location code', right_on='Country_Number')
```

```python
countries['Life Expectancy at Birth, both sexes (years)'] = pd.to_numeric(countries['Life Expectancy at Birth, both sexes (years)'], errors='coerce')
```

```python
countries[['Life Expectancy at Birth, both sexes (years)']].info()
```

```python
countries['Life Expectancy at Birth, both sexes (years)'].isna().sum()
```

```python
countries[['Life Expectancy at Birth, both sexes (years)']].describe()
```

```python
grouped_countries = countries.groupby(['Continent_Name', 'Year']).agg({'Life Expectancy at Birth, both sexes (years)': 'mean'})
grouped_countries = grouped_countries.pivot_table('Life Expectancy at Birth, both sexes (years)', index='Continent_Name', columns='Year')
grouped_countries
```

```python
colors = ['orange', 'yellow', 'blue', 'green', 'red', 'purple']
for continent, row in grouped_countries.iterrows():
    plt.plot(row)
plt.legend(grouped_countries.index)
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
```

```python
pivot_countries = countries[['Region, subregion, country or area *' , 'Year', 'Life Expectancy at Birth, both sexes (years)', 'Continent_Name']]
pivot_countries = pivot_countries.pivot_table('Life Expectancy at Birth, both sexes (years)', index=['Region, subregion, country or area *', 'Continent_Name'], columns='Year')
pivot_countries = pivot_countries.reset_index(level=[1])
pivot_countries
asia = pivot_countries[pivot_countries['Continent_Name'] == 'Europe'].drop('Continent_Name', axis=1)
for index, row in asia.iterrows():
    plt.plot(row)
```

```python
pivot_countries
```

```python
top_countries = pivot_countries[[2021.0, 2000.0]]
top_countries['Most increased'] = top_countries[2021.0] - top_countries[2000.0]
top_countries = top_countries[['Most increased']].sort_values('Most increased', ascending=False).iloc[:10]
top_countries.plot(kind='bar')
```

```python
incomes = all_data.iloc[1155: 1515,][['Year', 'Region, subregion, country or area *', 'Life Expectancy at Birth, both sexes (years)']].set_index('Region, subregion, country or area *')
incomes = incomes.pivot_table('Life Expectancy at Birth, both sexes (years)', index='Region, subregion, country or area *', columns='Year').sort_values(1950.0, ascending=False)
incomes
```

```python
for index, row in incomes.iterrows():
    plt.plot(row)
plt.legend(incomes.index)
```

```python
GDP = pd.read_csv('GDP.csv', skiprows=4)
GDP = GDP.drop(['Country Name', 'Indicator Name', 'Indicator Code'], axis=1)
```

```python
GDP
```

```python
GDP = GDP.melt(id_vars='Country Code', value_vars=set(GDP.columns) - set('Country Code'), var_name='Year', value_name='GDP_per_capita').sort_values('Country Code')
GDP.head()
```

```python
GDP['Year'] = pd.to_numeric(GDP['Year'], errors='coerce')
```

```python
GDP.info()
```

```python
continents_GDP = countries.merge(GDP, how='left', left_on=['Three_Letter_Country_Code', 'Year'], right_on=['Country Code', 'Year'])[['Continent_Name', 'GDP_per_capita', 'Year']]
```

```python
continents_GDP = continents_GDP[continents_GDP['GDP_per_capita'].notna()]
```

```python
continents_GDP = continents_GDP.groupby(['Continent_Name', 'Year']).agg({'GDP_per_capita': np.mean})
continents_GDP.head(20)
```

```python
continents_GDP = continents_GDP.pivot_table('GDP_per_capita', index='Continent_Name', columns='Year')
```

```python
continents_GDP
```

```python
for index, row in continents_GDP.iterrows():
    plt.plot(row)
plt.xlabel('Year')
plt.ylabel('GDP Per Capita')
plt.legend(continents_GDP.index)
```
