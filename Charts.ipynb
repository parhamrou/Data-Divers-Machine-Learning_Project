```python
q8=df[df['Region, subregion, country or area *']=='WORLD']
columns = ['Male Life Expectancy at Age 15 (years)', 'Female Life Expectancy at Age 15 (years)', 'Male Life Expectancy at Age 65 (years)', 'Female Life Expectancy at Age 65 (years)', 'Male Life Expectancy at Age 80 (years)', 'Female Life Expectancy at Age 80 (years)']
q8[columns] = q8[columns].apply(pd.to_numeric, errors='coerce', axis=1)
Male_Life_15=q8['Male Life Expectancy at Age 15 (years)'].mean()
Female_Life_15=  q8['Female Life Expectancy at Age 15 (years)'].mean()
Male_Life_65 =  q8['Male Life Expectancy at Age 65 (years)'].mean()
Female_Life_65 =   q8['Female Life Expectancy at Age 65 (years)'].mean()
Male_Life_80 =q8['Male Life Expectancy at Age 80 (years)'].mean()
Female_Life_80 = q8['Female Life Expectancy at Age 80 (years)'].mean()
ages=['15','65','80']
male=[Male_Life_15,Male_Life_65,Male_Life_80 ]
female=[Female_Life_15,Female_Life_65,Female_Life_80 ]
q8[columns].info()
```

```python
plt.bar(ages,male, label = 'male')

plt.bar(ages,female,bottom=male, label='female')

plt.legend();
```
